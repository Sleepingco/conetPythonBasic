```python
# • 위의 pivot() 메소드와 유사한 pivot_table() 이라는 메소드를 통해서 데이터를
# 재구조화하는 것도 가능하다.

# • 하지만 인덱스가 2개 이상인 경우 pivot() 메소드로는 재구조화가 안되며
# pivot_table()을 사용해야만 한다.

# • 마찬가지로 column이 2개 이상인 경우 pivot() 메소드로는 재구조화가 안되며 pivot_table()을
# 사용해야만 한다.

age = [22.0, 38.0, 26.0, 35.0, 35.0]        # 길이 5
sex = ['male','female','female','female','male']  # 길이 5
fare = [7.25, 71.2833, 7.9250, 53.1000, 8.0500]   # ✅ 길이 5로 수정 필요
cls = ['Third','First','Third','First','Third']   # "Thrid" 오타도 수정!
survived = [0, 1, 1, 1, 0]                        # 길이 5

df = pd.DataFrame({"age":age,"sex":sex,"cls":cls,"fare":fare,"survived":survived})

pdf3 = pd.pivot_table(df,
                      index = ["cls","sex"],
                      columns="survived",
                      values=["age","fare"])
pdf3
```

```python
# concat()
df_1 = pd.DataFrame({
    'A':['a10','a11','a12'],
    'B':['b10','b11','b12'],
    'C':['c10','c11','c12']},
    index=['가','나','다']
)
# concat()
df_2 = pd.DataFrame({
    'B':['b23','b24','b25'],
    'C':['c23','c24','C25'],
    'D':['d23','d24','d25'] },
    index=['다','라','마']
)
df_3 = pd.concat([df_1,df_2])
df_4 = pd.concat([df_1,df_2],join='inner')
df_3
df_4
```

```python
# merge
print('left outer \n',df_1.merge(df_2,how='left',on='B'))
print('right outer \n',df_1.merge(df_2,how='right',on='B'))
print('full outer \n',df_1.merge(df_2,how='outer',on='B'))
print('inner \n',df_1.merge(df_2,how='inner',on='B'))
# 두 데이터프레임에서는 B뿐만이 아니라 C 레이블도 두테이블에 동시에 나타난다
# 이럴 경우 왼쪽 테이블에서 가져온것은 C_x오른쪽에서 가져온 것은 C_y로 이름을 새로 붙여 테이블을 만든다
# 양쪽 테이블의 인덱스를 모두 사용하여 병함의키가 되는 방식은 아래와 같다
# outer 방식이므로 두 인덱스의 합집합이 새로운 인덱스가 된다
# inner 라면 교집합으로 다 행만 남게된다
print('outer \n',df_1.merge(df_2,how='outer',left_index=True,right_index=True))
```
